```python
import numpy as np
import pandas as pd
```

## 1. Bag of Words

```python
df = pd.DataFrame({'text':['people watch friends','friends watch friends','people write comment','friends write comment'],'output':[1,1,0,0]})
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
# vocab
cv.vocabulary_
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(['friends watch and write comment of friends']).toarray()
```

### Advantages :<br>
1. simple and intuitive
2. easy to implement 

### Disadvantages:<br>
1. sparsity
2. no fixed size
3. out of vocabulary
4. ordering is an issue

## 2.Bag of N-Grams

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

### Benefits<br>
1. Able to capture semantic of sentence
2. easy to implement
3. easy to intuit

### Disadvantages<br>
1. Dimension of vocaublary increases
2. Slows down the algo
3. No solution for out of vocabulary
